```python
import numpy as np

vec_list_eng = []
for year in range(1800,2000,10):
    vec = np.load(f"english_sgns/{year}-w.npy") 
    vec_list_eng.append(vec)

count_list_eng = []
for vec in vec_list_eng:
    count = 0
    for i in range(len(vec)):
        if np.count_nonzero(vec[i]) > 0:
            count += 1
    count_list_eng.append(count)
print(count_list_eng)



vec_list_ger = []
for year in range(1800,2000,10):
    vec = np.load(f"german_sgns/{year}-w.npy") 
    vec_list_ger.append(vec)

count_list_ger = []
for vec in vec_list_ger:
    count = 0
    for i in range(len(vec)):
        if np.count_nonzero(vec[i]) > 0:
            count += 1
    count_list_ger.append(count)
print(count_list_ger)



vec_list_fre = []
for year in range(1800,2000,10):
    vec = np.load(f"french_sgns/{year}-w.npy") # changeable
    vec_list_fre.append(vec)

count_list_fre = []
for vec in vec_list_fre:
    count = 0
    for i in range(len(vec)):
        if np.count_nonzero(vec[i]) > 0:
            count += 1
    count_list_fre.append(count)
print(count_list_fre)
```

```python
import pandas as pd

years = []
for year in range(1800,2000,10):
    years.append(year)

df = pd.DataFrame(list(zip(years, count_list_eng, count_list_ger, count_list_fre)), columns =['Decade', 'English_Words', 'German_Words', 'French_Words'])
df
```

```python
ax = df.plot(x="Decade", y=["English_Words", "German_Words", "French_Words"], kind="bar", figsize=(25,10), title='Amount of Words with Embeddings per Decade');
ax.set_xlabel("\nDecade", fontsize=12);
ax.set_ylabel("Number of Words\n", fontsize=12);
```
